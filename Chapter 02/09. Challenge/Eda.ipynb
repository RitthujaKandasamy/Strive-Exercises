```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Check the unique values

unique_values = []
for i in df.columns:
    unique_count = df[i].value_counts().count()
    unique_values.append([i,unique_count])
```

```python
print(unique_values)
```

```python
# Exploring the data

# Plot the age
sns.set(rc={"figure.figsize":(12,8)})
sns.displot(df['age'].dropna())
plt.show()
```

```python
# Plot the sex

column = {'0': 'female', '1': 'male'}
sns.countplot(x='sex',data=df,palette='Set3')
plt.show()
```

```python
# Correlation Matrix

matrix = df.corr().round(2)
sns.set(rc={"figure.figsize":(20,10)})
sns.heatmap(matrix, annot=True, cmap="YlOrBr")
plt.show()
```

```python
df.corr()["output"].abs().sort_values().plot.barh()

```

```python
df.corr()["output"].sort_values().plot.barh()
```

```python
plt.figure(figsize = (10,6))
plt.xticks(size = 20,color ='grey')
plt.tick_params(size = 12,color ='grey')
plt.title('Finding Null Values Using Heatmap\n',color ='grey',size = 30)
sns.heatmap(df.isnull(),
            yticklabels = False,
            cbar = False,
            cmap = 'PuBu_r',
            )
plt.show()
```

```python
numeric_var = ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak']

for i in numeric_var:
    plt.figure(figsize = (10, 6))
    sns.distplot(df[i], hist_kws = dict(linewidth = 1, edgecolor = "k"), bins = 20)
    
    plt.title(i)
    plt.xlabel(i)

    
    plt.tight_layout()
    plt.show()

```
