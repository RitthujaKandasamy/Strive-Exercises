#### Checking the number of unique values in the dataset to determine the categorical and numerical column 

```python
unique_values = []
for i in data.columns:
    unique_count = data[i].value_counts().count()
    unique_values.append([unique_count])
pd.DataFrame( unique_values,   index= data.columns, columns= ['Unique value count'] )
```

```python
#the categorical variables are the indexs with fewer count values 
cat_var = ['target', 'f_30', 'f_29', 'f_18','f_17','f_16','f_15','f_14','f_13','f_12','f_11','f_10','f_09','f_08','f_07']
num_var = ['f_28','f_26','f_25','f_24','f_23','f_22','f_21','f_20','f_19','f_06','f_05','f_04','f_03','f_02','f_01','f_00','id']
```

## Correlation Matrix for numerical data

```python
plt.figure(figsize=(12, 12))
sns.heatmap(data[num_var + ['target']].corr(), center=0, annot=True, fmt='.2f', cmap='YlGnBu')
plt.show()
```

## Correlation Matrix

```python
plt.figure(figsize = (18, 15))
data.corr()["target"].abs().sort_values().plot.barh()
plt.title("Correlation of features in the Dataset", fontsize = 25)
plt.xlabel("Correlation", fontsize = 20)
plt.ylabel("Features", fontsize = 20)
```

 No good correlation features are: f_04, f_03, f_12, f_06, f_17, f_07, f_18

### Univariate Analysis

```python
sns.set(rc={"figure.figsize":(10,10)})
sns.countplot(x ='target', data = data,hue="target")
plt.title("Distribution of the target variable", fontsize = 20)
plt.show()
```

## Categorical data visualization

```python
title_font = {"family" : "arial", "color" : "darkred", "weight" : "bold", "size" : 15}
axis_font = {"family" : "arial", "color" : "darkblue", "weight" : "bold", "size" : 13}

for i in cat_var :
    plt.figure(figsize = (8, 5))
    sns.countplot(i, data = data[cat_var], hue = "target")
    
    plt.title(i + " - target", fontdict = title_font)
    plt.ylabel("Target", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

## Numerical data visualization

```python
for i in num_var:
    plt.figure(figsize = (8, 6))
    sns.distplot(data[i], hist_kws = dict(linewidth = 1, edgecolor = "k"), bins = 20)
    
    plt.title(i)
    plt.xlabel(i)

    
    plt.tight_layout()
    plt.show()
```
