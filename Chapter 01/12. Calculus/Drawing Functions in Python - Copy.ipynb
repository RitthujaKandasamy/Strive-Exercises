```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

## Exercises

1. Plot the absolute value of x and its derivative. Choose an interval to have a nice looking shape.

```python
def absolute_value(x):
    return abs(x)

x = np.linspace(-5, 1, 100)
y = absolute_value(x)
plt.plot(x, y, label = "Absolute Value")

# dx is can take any value, but mostly we always use dx = 1e-6

absolute_derivative = derivative(absolute_value, x, dx = 6e-15)
plt.plot(x, absolute_derivative, label = "Derivative")
plt.title("Absolute Value in derivative")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()

```

2. Plot the function $x\log(x)$ and its derivative.

```python
def xlog(x):
    return x * np.log(x)

x = np.linspace(10, 100, 10000)
y = xlog(x)
plt.plot(x, y, label = "Log(x)")

log_derivative = derivative(xlog, x, dx = 1e-9)
plt.plot(x, log_derivative, label = "Derivative")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

3. Plot the function $e^{x}\log(x)+5$ and its derivative.

```python
def elog(x):
    return np.exp(x) * np.log(x) + 5

x = np.linspace(5, 6, 10000)
y = elog(x)
plt.plot(x, y, label = "Log(x)", linestyle ='--')

# value dx is changeable, it is used for giving the smaller interval in line
# without dx also we can work

log_derivative = derivative(elog, x)
plt.plot(x, log_derivative, label = "Derivative")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

4.(Harder) Plot the function $\sin(\frac{1}{2}x)+2$ in the interval \[0, $2\pi$\]. Plot in the same graph both the function and the tangent line to the points in which the function has derivative = 0.

```python
def sin_value(x):
    return np.sin((1/2) * x) + 2 
    
x = np.linspace(0, 2 * np.pi, 1000)
y = sin_value(x)
plt.plot(x, y, label = "Sin Value")

sin_derivative = derivative(sin_value, x, dx = 1e-6)

# to draw tangent line in the x and y axis
# we take for loop

x1 = 0
for i in range(0, len(sin_derivative)):
    if -0.0008 <= sin_derivative[i]  <= 0.0008:       # inbetween -0.0008 to 0.0008 lies because the sin_derivative is near to 0
        x1 = x[i]

y1 = sin_value(x1)

plt.axhline(y = y1, label = "Derivative value = 0", color = "orange")
plt.title("Tangent Value in derivative = 0")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()

```
