```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image\Plot.png")
```

```python
type(img)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize= (15, 10))
plt.imshow(gray_img, cmap = "gray")
```

```python
retval, dst = cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

plt.figure(figsize= (15, 10))
plt.imshow(dst, cmap= "gray")
```

# Contours

```python
contours, hierachy = cv2.findContours(dst, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
type(contours)
```

```python
drawing = cv2.drawContours(gray_img, contours, -1, (0, 255, 0), 7)

plt.figure(figsize=(15, 10))
plt.imshow(drawing, cmap= "gray")
```

```python
sorted_contours = sorted(contours, key= cv2.contourArea, reverse= True)
drawing2 = cv2.drawContours(img, sorted_contours[0: 26], 0, (0, 255, 0), 1)

plt.figure(figsize= (15, 10))
plt.imshow(drawing2, cmap = "gray")
```

# Bounding Rectangle

```python
clean_contours = sorted_contours[0: 26]
x, y, w, h = cv2.boundingRect(clean_contours[0])
```

```python
rect = cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

plt.figure(figsize= (15, 10))
plt.imshow(rect)
```

```python
zero = img[y:y+h, x:x+w]

plt.figure(figsize= (15, 10))
plt.imshow(zero)
plt.savefig('./c0.png')
```
