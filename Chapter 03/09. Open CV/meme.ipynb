```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("image\\thanos.jpg")
```

### Create memes

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 10))
plt.imshow(rgb_img)
```

```python
img_copy = rgb_img.copy()
plt.figure(figsize= (15, 10))
cv2.putText(img_copy, 'THANOS MEME', (450, 100), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 127))
cv2.putText(img_copy, 'IMPOSSIBLE', (200, 500), cv2.FONT_HERSHEY_TRIPLEX, 2, (225, 0, 0))
plt.imshow(img_copy)
```
