```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("image\\thanos.jpg")
```

### Create memes

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 10))
plt.imshow(rgb_img)
```

```python
img_copy = rgb_img.copy()
plt.figure(figsize= (15, 10))
cv2.putText(img_copy, 'THANOS MEME', (450, 100), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 0), 3)
cv2.putText(img_copy, 'IMPOSSIBLE', (200, 500), cv2.FONT_HERSHEY_TRIPLEX, 2, (255, 255, 255), 3)
plt.imshow(img_copy)
```

```python
def meme_gen(image, text_up = str, text_down = str):

    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    font = cv2.FONT_HERSHEY_COMPLEX
   
    img1 = cv2.putText(img, text_up, (30, 250), font, 1.7, (255, 255, 255), 2)
    meme = cv2.putText(img1, text_down, (320, 1750), font, 2, (255, 255, 255), 3)
    plt.figure(figsize = (10, 8))
    plt.imshow(meme)

meme_gen('image\cartoon-wallpaper-ixpap.png', 'Yarum ennai parkka mutiyatu', 'Angry comes only !')
```

### QR code Detect

```python
qr_img = cv2.imread('image\qr.jpg')
plt.figure(figsize = (12, 8))
plt.imshow(qr_img)
```

```python
decoder = cv2.QRCodeDetector()

data, points, _ = decoder.detectAndDecode(qr_img)
```

```python
if points is not None:
    print('Decoded data: ', data)
    points = points[0]
    for i in range(len(points)):
        pt1 = [int(val) for val in points[i]]
        pt2 = [int(val) for val in points[(i + 1) % 4]]
        cv2.line(qr_img, pt1, pt2, color = (0, 255,0), thickness = 3)
 
# cv2.imshow('Detected QR code', qr_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
