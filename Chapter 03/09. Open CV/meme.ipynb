```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread("image\\thanos.jpg")
```

### Create memes

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_grey = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(15, 10))
plt.imshow(rgb_img)
```

```python
#copying the image
copy = rgb_img.copy()
font = cv2.FONT_HERSHEY_SIMPLEX

# setting the 2 phrases taht should be added to the picture
text1 = 'This is not funny,'
text2 = 'but a Meme!'

# adding the phrases to the picture 
if len(text1) > 0 and len(text2) >0:
    textsize1 = cv2.getTextSize(text1, font, 2, 2)
    textsize2 = cv2.getTextSize(text2, font, 2, 2)
    textX1 = int((img_grey.shape[1] - textsize1[0][0]) / 2)
    textY1 = int(img.shape[0] - (img.shape[0] - 2 * textsize1[1]))
    textX2 = int((img_grey.shape[1] - textsize2[0][0]) / 2)
    textY2 = int((img.shape[0] - textsize2[1]))
    cv2.putText(copy, text1, (textX1, textY1), font, 2, (0, 0, 0), 3)
    cv2.putText(copy, text2, (textX2, textY2), font, 2, (0, 0, 0), 3)
    


# showing the Meme
plt.figure(figsize=(10, 8))
plt.imshow(copy)
```

```python
img_copy = rgb_img.copy()
plt.figure(figsize= (15, 10))
cv2.putText(img_copy, 'THANOS MEME', (450, 100), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 0), 3)
#cv2.putText(img_copy, 'THANOS MEME', (450, 100), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 10)
cv2.putText(img_copy, 'IMPOSSIBLE', (200, 500), cv2.FONT_HERSHEY_TRIPLEX, 2, (255, 255, 255), 3)
plt.imshow(img_copy)
plt.savefig('./thanos_meme.png')
```

```python
def meme_gen(image, text_up = str, text_down = str):

    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    font = cv2.FONT_HERSHEY_COMPLEX
   
    img1 = cv2.putText(img, text_up, (30, 250), font, 1.7, (255, 255, 255), 2)
    meme = cv2.putText(img1, text_down, (320, 1750), font, 2, (255, 255, 255), 3)
    plt.figure(figsize = (10, 8))
    plt.imshow(meme)

meme_gen('image\cartoon-wallpaper-ixpap.png', 'Yarum ennai parkka mutiyatu', 'Angry comes only !')
```

# Bitwise

```python
# only odd numbers 
background = np.zeros([600, 600 ,3])

plt.figure(figsize=(10,10))
plt.imshow(background.astype('uint8'))
```

```python
cv2.circle(background, (200, 200), 100, (255, 255, 255), cv2.FILLED)

plt.figure(figsize= (10, 10))
plt.imshow(background.astype('uint8'))
```

```python
cv2.rectangle(background, (200, 250), (400,400), (255, 255, 255), cv2.FILLED)

plt.figure(figsize=(10,10))
plt.imshow(background.astype('uint8'))
```

```python
cv2.line(background, (250, 410), (0, 0), (255, 255, 255), 1)

plt.figure(figsize=(10,10))
plt.imshow(background.astype('uint8'))
```
