```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("image\\thanos.jpg")
```

### Create memes

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 10))
plt.imshow(rgb_img)
```

```python
img_copy = rgb_img.copy()
plt.figure(figsize= (15, 10))
cv2.putText(img_copy, 'THANOS MEME', (450, 100), cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 0), 3)
cv2.putText(img_copy, 'IMPOSSIBLE', (200, 500), cv2.FONT_HERSHEY_TRIPLEX, 2, (255, 0, 0), 3)
plt.imshow(img_copy)
```

### Create Annotation

```python
ann_img = cv2.imread("img\img\dog.jpg")
```

```python
ann_rgb_img = cv2.cvtColor(ann_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 10))
plt.imshow(ann_rgb_img)
```

```python
cv2.rectangle(ann_rgb_img,(150, 50), (600, 600), (255, 0, 0), 4)
cv2.rectangle(ann_rgb_img, (150, 50),(300, 0), (255, 0, 0), cv2.FILLED)
cv2.putText(ann_rgb_img,'Dog', (190, 30), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,0))

plt.figure(figsize=(15, 10))
plt.imshow(ann_rgb_img)
```
