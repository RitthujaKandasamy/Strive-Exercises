```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

### Create hsv and image source

```python
hsv = cv2.imread("img\img\hsv.png")
img = cv2.imread("img\img\m.jpg")
```

```python
img
```

```python
hsv_img = cv2.cvtColor(hsv, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 10))
plt.imshow(hsv_img)
```

```python
hsv_m_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.figure(figsize=(15, 10))
plt.imshow(hsv_m_img)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 10))
plt.imshow(rgb_img)
```

### Take out only Blue with mask

```python
lower_blue = np.array([90, 50, 80])
upper_blue = np.array([150, 255, 255])
```

```python
mask_blue = cv2.inRange(hsv_m_img, lower_blue, upper_blue)
res_blue = cv2.bitwise_and(rgb_img, rgb_img, mask = mask_blue)

plt.figure(figsize=(15, 10))
plt.imshow(res_blue)
```

### Take out only Green from Image

```python
lower_green = np.array([35, 30, 20])
upper_green = np.array([85, 255, 255])
```

```python
mask_green = cv2.inRange(hsv_m_img, lower_green, upper_green)
res_green = cv2.bitwise_and(rgb_img, rgb_img, mask = mask_green)

plt.figure(figsize=(15, 10))
plt.imshow(res_green)
```

### Showing Both Blue and Green in Image

```python
lower_gb = np.array([40, 50, 0])
upper_gb = np.array([130, 255, 255])
```

```python
mask_gb = cv2.inRange(hsv_m_img, lower_gb, upper_gb)
res_gb = cv2.bitwise_and(rgb_img, rgb_img, mask = mask_gb)

plt.figure(figsize=(15, 10))
plt.imshow(res_gb)
```
