```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,15))
plt.imshow(img)
```

```python
#Reason for not matched color:
#openCV      B G R
#MAtplotlib  B G R
```

```python
#inorder to fix that:
rgb_img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)#Converts color
rgb_img
```

```python
plt.figure(figsize=(20,15))
plt.imshow(rgb_img)
#Now colors will match
```

```python
# what about gray image?
gray_img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)#Converts color

cv2.imshow('MINIONS', gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1) #Questions in here
```

```python
plt.figure(figsize=(10,8))
plt.imshow(gray_img, cmap='gray')#try it without cmap
```
